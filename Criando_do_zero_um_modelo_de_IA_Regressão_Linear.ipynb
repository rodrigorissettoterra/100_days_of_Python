# Bibliotecas

```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

# Dataset

```python
housing = datasets.fetch_california_housing()
```

Informações sobre o dataset utilizado

```python
print(housing.DESCR)
```

# Preparação dos dados

Convertendo o dataset em dataframe

```python
df = pd.DataFrame(housing.data, columns=housing.feature_names)
```

```python
df
```

Definindo o que deverá ser previsto (target) com o algoritmo

```python
dfTarget = pd.DataFrame(housing.target, columns = housing.target_names)
```

```python
dfTarget
```

Definindo o que é X de treino, X de teste, Y de treino e Y de teste

```python
X = df
```

```python
Y = dfTarget
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=123)
```

Escalonando os valores de X

```python
scaler = StandardScaler()
```

```python
X_train_Scaled = scaler.fit_transform(X_train)
X_test_Scaled = scaler.fit_transform(X_test)
```

# ALgoritmo de Regrassão Linear

```python
reg = LinearRegression().fit(X_train_Scaled, y_train)
```

```python
results = reg.predict(X_test_Scaled)
```

```python
results
```

Score do modelo

```python
reg.score(X_test_Scaled, y_test)
```
